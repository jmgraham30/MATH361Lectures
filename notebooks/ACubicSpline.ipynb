```julia
using Polynomials, Plots, LaTeXStrings
default(markersize=3,linewidth=1.5);
```

```julia
f(x) = x^4;
nodes = [0.0,1.0,2.0];
h = 1.0;
y_nodes = @. f(nodes);
```

```julia
p = plot(f,0,2,label="f")
p = scatter!(p,nodes,y_nodes,label="nodes",legend=:topleft)
```

```julia
function s(x)
    if x >= 0.0 && x <= 1.0
        s_x = 7/2*x^3 - 5/2*x
    else
        s_x = -7/2*x^3 + 21*x^2 - 47/2*x + 7
    end

    return s_x
end
```

```julia
x_val = 0.0:0.01:2.0;
s_y = @. s(x_val);
p = plot!(p,x_val,s_y,label="s")
```

```julia
error = @. abs(f(x_val) - s(x_val))
```

```julia
max_err = maximum(error)
```
