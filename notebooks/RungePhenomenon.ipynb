```julia
using Polynomials, Plots, LaTeXStrings
default(markersize=3,linewidth=1.5);
```

```julia
f_runge(x) = 1/(1+x^2);
p = plot(f_runge,-5.0,5.0,legend=false)
```

```julia
function build_lagrange(x;nodes,f)
    
    y = f.(nodes);
    np1 = length(nodes);
    pn = 0.0;
    
    for k=1:np1
        xt = copy(nodes);
        deleteat!(xt,k)
        num = x .- xt;
        dom = nodes[k] .- xt;
        Q = (num)./(dom);
        L = prod(Q);
        pn += y[k]*L;
    end
    
    return pn
end   
```

```julia
nodes1 = [-5,5];
p1(x) = build_lagrange(x;nodes=nodes1,f=f_runge)

plot!(p,p1,-5.0,5.0)
```

```julia
nodes2 = [-5,0.0,5];
p2(x) = build_lagrange(x;nodes=nodes2,f=f_runge)

plot!(p,p2,-5.0,5.0)
```

```julia
n = 3
dx = (5.0 + 5.0)/n
nodes3 = [x for x in -5.0:dx:5.0];
p3(x) = build_lagrange(x;nodes=nodes3,f=f_runge)

plot!(p,p3,-5.0,5.0)
```
